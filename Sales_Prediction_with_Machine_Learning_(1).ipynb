```python
import pandas as pd
data = pd.read_csv("/content/Advertising.csv",index_col=0)
data
```

```python
data.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(15,6))
sns.lineplot(data=data,x=data['TV'],y=data['Sales'])
```

```python
plt.figure(figsize=(15,6))
sns.lineplot(data=data,x=data['Radio'],y=data['Sales'])
```

```python
plt.figure(figsize=(15,6))
sns.lineplot(data=data,x=data['Newspaper'],y=data['Sales'])
```

```python
data.describe()
```

```python
x = data.drop(['TV','Radio','Newspaper'],axis=1)
y = data['Sales']
```

```python
plt.figure(figsize=(15,6))
plt.plot(x,y)
plt.title("Sales Rate")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
```

```python
rfr.fit(x_train,y_train)
```

```python
y_pred = rfr.predict(x_test)
```

```python
accuracy = rfr.score(x_test,y_test)
print("Accuracy:",accuracy*100)
```
